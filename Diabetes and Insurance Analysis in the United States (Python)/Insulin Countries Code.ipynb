Health Equity: An Examination of Insulin Access

This code aims to visualize the price of insulin given countries with universal healthcare/ a dominant public health care system compared to insulin prices of the US. 

```python
import pandas as pd
import matplotlib.pyplot as plt
country_data = 'data-x2xEg.csv'
```

```python
def clean_data(filename):
    insulin_countries = pd.read_csv(filename)
    # rename a column
    insulin_countries.rename(columns = {'X.1':'Years'}, inplace = True)
    # fill in empty year data
    insulin_countries['Switzerland'] = insulin_countries['Switzerland'].fillna(method='bfill')
    insulin_countries['United Kingdom'] = insulin_countries['United Kingdom'].fillna(method='bfill')
    
    return insulin_countries
```

```python
def plot(data):
    plt.figure(figsize = (8,8), dpi = 400)
    plt.scatter(data.Years, data['United States'], marker = '.', color = 'r', label = 'US insulin prices')
    plt.scatter(data.Years, data.Australia, marker = '.', color = 'b', label = 'Aus insulin prices')
    plt.scatter(data.Years, data.Austria, marker = '.', color = 'g', label = 'Austria insulin prices')
    plt.scatter(data.Years, data.Belgium, marker = '.', color = 'pink', label = 'Belgium insulin prices')
    plt.scatter(data.Years, data.Canada, marker = '.', color = 'black', label = 'Canada insulin prices')
    plt.scatter(data.Years, data.France, marker = '.', color = 'orange', label = 'France insulin prices')
    plt.scatter(data.Years, data.Japan, marker = '.', color = 'yellow', label = 'Japan insulin prices')
    plt.scatter(data.Years, data.Sweden, marker = '.', color = 'brown', label = 'Sweden insulin prices')
    plt.scatter(data.Years, data.Switzerland, marker = '.', color = 'grey', label = 'Switzerland insulin prices')
    plt.scatter(data.Years, data['United Kingdom'], marker = '.', color = 'purple', label = 'United Kingdom insulin prices')
    plt.plot(data.Years, data['Comparable Country Average'], marker = '.', color = 'hotpink', label = 'Average Country insulin prices')
    plt.legend(loc = 'upper left', fontsize = '8')
    plt.xlabel('Years')
    plt.ylabel('Price')
    plt.title('Country Insulin Prices over the Years')
    plt.grid()
    plt.savefig('insulin_countries.png')
    plt.show()
```

```python
def main():
    insulin_data = clean_data(country_data)
    print(insulin_data.head())
    insulin_plot = plot(insulin_data)
    
if __name__ == '__main__':
    main()
```
